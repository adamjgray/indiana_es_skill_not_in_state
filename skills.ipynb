```ruby
require "faraday"
require "nokogiri"
require "open-uri"
require "json"
require "awesome_print"
```

```ruby
url = "https://www.robotevents.com/api/seasons/129/skills?post_season=1&grade_level=Elementary"
skills = JSON.parse(Faraday.get(url).body)
indiana_skills = skills.select { |s| s["team"]["eventRegion"] == "Indiana" }
```

```ruby
registered = []
competition = "https://www.robotevents.com/robot-competitions/vex-iq-challenge/RE-VIQC-19-0988.html"
doc = Nokogiri::HTML(URI.open(competition))
doc.at("#data-table").search("tr").each { |tr| registered << tr.search("a").text }
```

```ruby
(indiana_skills.map { |s| s["team"]["team"] } - registered).each.with_index(1) { |t, i| puts "#{i}: #{t}" }
```
